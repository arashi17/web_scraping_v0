```python
url = 'https://www.sportsbet.com.au/betting/basketball-us'
headers = {'User-Agent': 'Mozilla/5.0 (Linux; Android 7.0; SM-G930V Build/NRD90M) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/59.0.3071.125 Mobile Safari/537.36'}
```

```python
import bs4 as bs
import requests
import pprint
import json
```

```python
sauce = requests.get(url, headers=headers)
soup = bs.BeautifulSoup(sauce.text, 'lxml')
```

```python
scripts = soup.findAll('script')
```

```python
script_text = scripts[2].text
print(script_text[0:34])
script_json = json.loads(script_text[34:])
```
