```python
import bs4 as bs
import requests
import pprint
import json
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/71.0.3578.98 Safari/537.36'}
```

```python
url_nba = 'https://www.oddschecker.com/au/basketball/nba'
```

```python
sauce = requests.get(url_nba, headers=headers)
```

```python
print(sauce)
```

```python
soup = bs.BeautifulSoup(sauce.text, 'lxml')
```

```python
print(soup)
```

```python
scripts = soup.findAll('script', {'id':'initial-data'})
```

```python
script_text = scripts[0].text
```

```python
print(type(script_text))
```

```python
script_json = json.loads(script_text)
```

```python
index = 0
```

```python
pprint.pprint(script_json['card']['matches'][0]['cards'][0]['data'][index]['bets'][0])
```

```python
pprint.pprint(script_json['card']['matches'][0]['cards'][0]['data'][index]['bets'][1])
```
